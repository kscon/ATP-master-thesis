```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib as mpl
```

## Prepare Data

```python
filepath = '../output/multiple_new_trains/directory_execution_time_comparison_'
filepath_suffix = '_ten_it_mip_it_mip_gl.csv'
#file1 = 'directory_execution_time_comparison_10b_ten_it_mip_it_mip_gl.csv'
#file2 = 'directory_execution_time_comparison_20b_ten_it_mip_it_mip_gl.csv'
#file3 = 'directory_execution_time_comparison_30b_ten_it_mip_it_mip_gl.csv'

files = ['10b','20b','30b', '40b', '50b']


list_df = {}
df_files = pd.DataFrame()
for f in files:
    df_tmp = pd.read_csv(filepath+f+filepath_suffix, index_col = 0, header=[0,1,2])
    list_df[f] = df_tmp.copy()
    df_files = pd.concat([df_files, df_tmp])


"""df_b = df_b.T
res = list(map(eval, df_b.index))
res = [(a,str(b),c) for (a,b,c) in res]
df_b = df_b.T
df_b.columns = pd.MultiIndex.from_tuples(res)
df_b.columns"""
df_files.head()
```

only take complete execution times of mip and ten

```python
df2 = df_files.drop(['ten_creation_time', 'ten_edge_deletion_time', 'ten_pathfinding_time',
              'mip_creation_time', 'mip_solving_time'], axis=1, level=2)

df2.head()
#df = df.loc[:, ['file_path', 'mip_execution_time', 'ten_execution_time']]
#df.head(15)
```

```python
df3 = df2.copy()
df3.columns = df3.columns.droplevel(2)
#df3.columns = df3.columns.to_flat_index()
df3.columns = df3.columns.get_level_values(0) + '_' +  df3.columns.get_level_values(1)
df3.reset_index(inplace=True)
df3['filename'] = df3['index']
df3 = df3.drop('index', axis=1)
df3 = df3[[s for s in sorted(df3.columns)]]
df3.head()
```

```python
ten_iterations = [s for s in df3.columns if 'ten_it' in s]
mip_iterations = [s for s in df3.columns if 'mip_it' in s]
#for s in ten_iterations: 
#    print(df3[s]) 
df3['ten_it_combined_time'] = sum(df3[s].fillna(0) for s in ten_iterations)
df3['mip_it_combined_time'] = sum(df3[s].fillna(0) for s in mip_iterations)
df3['mip_gl_combined_time'] = df3['mip_gl_0']
df3.head()
```

### average runtimes of all models grouped by setting

```python
#### average running time 

df_avg_runtimes = df3.copy()
test_classes = [f[0:14] for f in df_avg_runtimes['filename']]
#df_avg_runtimes['class'] = 
#df_avg_runtimes.head()
test_classes
df_avg_runtimes = df_avg_runtimes[['ten_it_combined_time','mip_it_combined_time', 'mip_gl_combined_time']]
df_avg_runtimes['test_classes'] = test_classes
df_mean = df_avg_runtimes.groupby(by='test_classes').mean()
df_mean = df_mean.reset_index()
# extra hustle 
mask2 = ['10et' in fp for fp in df_mean['test_classes']]
mask3 = ['10et' not in fp for fp in df_mean['test_classes']]
df_mean = pd.concat([df_mean[mask3], df_mean[mask2]])
df_mean.head(10)
```
